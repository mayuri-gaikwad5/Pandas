# Handling Missing Data

## dropna & fillna

```python
import pandas as pd
var = pd.read_csv("C:\\Users\\2309m\\Desktop\\MY WORK\\Github\\Pandas\\data (1).csv")
var
```

```python
var.dropna() #drops null values
var.dropna(axis=1) # drops a column which ha null values
```

```python
var.dropna(how="any")  
```

```python
var.dropna(how="all")# removes the row which has all null values in this row 30 is a example
```

## subset parameter

```python
var.dropna(subset=['Date'] )# all the null values present in column Date are removed
```

# how to fill value

```python
var.fillna("python") # fills the null values with word python
```

```python
var.fillna({'Date':'python','Pulse':'dead'}) # refer row 96 for changes
```

```python
var.fillna(method='ffill') # fills the null values with forward data
var.fillna(method='bfill') # fills the null values with backward data
```

```python
var.fillna(method='bfill',axis=1) # along columns
```

```python
var.fillna('hiiiiiiii',limit=1) # a column has only one null value
```
