# Interpolate

```python
import pandas as pd
var = pd.read_csv("C:\\Users\\2309m\\Desktop\\MY WORK\\Github\\Pandas\\data (1).csv")
var
```

```python
var = var.infer_objects(copy=False)  # This safely converts columns like 'object' to 'float', 'int', etc.
var = var.interpolate(method='linear')  # or with other parameters
# var.interpolate(method='linear',axis=0)
var
```

```python
var.replace(1,method="ffill",limit=3,inplace=True)
var.replace(limit_direction="both",limit=2,inplace=True)
```
